```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import RandomizedSearchCV, cross_val_score, TimeSeriesSplit
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
import sklearn.metrics
import math



df = pd.read_csv('train_cleaned_wide.csv')
df = df.drop(df.columns[0], axis=1)
```

```python
df_train = df
y_train = df_train['TotalAmountofAssistance'].values
df_test = pd.read_csv('test_data_wide.csv')
y_test = df_test['TotalAmountofAssistance'].values

# Feature transformation for the entire train set and test set
enc = OneHotEncoder(handle_unknown='error', sparse=False)
scaler = StandardScaler()

nominal = ['State']
numerical = ['CO2_Emissions', 'TotalNumberofInvestments', 'BDFDB', 'BDPRP', 'BFFDB', 'BFPRP', 'CLPRB', 'CLPRK', 'CLPRP', 'COPRK', 'EMFDB', 'ENPRP', 'GETCB', 'HYTCB', 'NCPRB', 'NGMPB', 'NGMPK', 'NGMPP', 'NUETB', 'PAPRB', 'PAPRP', 'REPRB', 'SOTCB', 'TEPRB', 'TETCB', 'WDEXB', 'WDPRB', 'WDTCB', 'WSTCB', 'WWPRB', 'WYTCB']

# Train set
X_nominal = enc.fit_transform(df_train[nominal])
nominal_cols = enc.get_feature_names_out(nominal)

X_num = df_train[numerical].copy()
for col in numerical:
    X_num[col] = scaler.fit_transform(df_train[[col]])

X_train = np.concatenate((X_nominal, X_num), axis=1)
print(X_train.shape)

# Test set
X_nominal = enc.fit_transform(df_test[nominal])
nominal_cols = enc.get_feature_names_out(nominal)

X_num = df_test[numerical].copy()
for col in numerical:
    X_num[col] = scaler.fit_transform(df_test[[col]])

X_test = np.concatenate((X_nominal, X_num), axis=1)
print(X_test.shape)
```

```python
# Train KNN model model and predict the test set
KNN_RMSEs = []
Ks = range(1,31,2)
for k in Ks:
  KNN = KNeighborsRegressor(n_neighbors=k)
  # Train the model on training data
  KNN_fitted = KNN.fit(X_train, y_train)

  # Use the forest's predict method on the test data
  predictions = KNN_fitted.predict(X_test)

  # Calculate the absolute errors
  mse = sklearn.metrics.mean_squared_error(y_test, predictions)

  KNN_RMSEs.append(math.sqrt(mse))

sns.lineplot(x=Ks, y=KNN_RMSEs, marker='+', label='RMSE')
plt.title('K vs. RMSE')
plt.xlabel('K')
plt.ylabel('RMSE')

for i in range(len(Ks)):
    plt.text(x = Ks[i], y = KNN_RMSEs[i] + 0.001, s=f'K = {Ks[i]}', color = 'blue', size = 'large')
plt.show()
```

```python
# Train Random Forest model and predict the test set
RFC_RMSEs = []
for i in range(10):
  RFC = RandomForestRegressor(n_estimators=100, oob_score=True)

  # Train the model on training data
  RF_fitted = RFC.fit(X_train, y_train)

  # Use the forest's predict method on the test data
  predictions = RF_fitted.predict(X_test)

  # Calculate the mean squared error
  mse = sklearn.metrics.mean_squared_error(y_test, predictions)
  
  RFC_RMSEs.append(math.sqrt(mse))

print(np.mean(RFC_RMSEs))
```

```python
# Plot feature importances
enc_cols = np.concatenate((nominal_cols, numerical))
RFC.fit(X_train, y_train)
importances = RFC.feature_importances_
importances = pd.DataFrame(importances, index=enc_cols, columns=['Values'])
importances = importances.sort_values(by='Values', ascending=False).head(20)
fig, ax = plt.subplots(figsize=(15,8))
sns.barplot(y=importances['Values'], x=importances.index, order=importances.sort_values(by='Values', ascending=False).index, ax=ax)
ax.set_ylabel('Mean decrease in impurity')
sns.set(font_scale=2)
plt.xticks(rotation=90)
plt.show()
```

```python
# XGBoost hypaerparameter tuning to find the best hyperparameteters combination
param_grid = {
    'max_depth': [4, 5, 6],
    'min_child_weight': [0.2991],
    'n_estimators': [1200, 1300, 1400],
    'gamma': [38.78],
    'learning_rate': [0.004, 0.005, 0.006],
    'subsample': [1],
    'colsample_bytree': [0.1, 0.3, 0.01]
}
cv = TimeSeriesSplit(n_splits=4).split(X_train)
XGB = XGBRegressor(objective ='reg:squarederror',)    
grid_search = RandomizedSearchCV(XGB, param_grid, scoring="neg_mean_squared_error", n_iter = 2000, cv=cv, n_jobs=-1)
grid_result = grid_search.fit(X_train, y_train)
```

```python
# Train XGBoost model and predict the test set
XGB = XGBRegressor(objective ='reg:squarederror', n_estimators=1300, subsample=1, min_child_weight=0.2991, max_depth=4, learning_rate=0.006, gamma=38.78, colsample_bytree=0.15)

# Train the model on training data
XGB_fitted = XGB.fit(X_train, y_train)

# Use the forest's predict method on the test data
predictions = XGB_fitted.predict(X_test)

# Calculate the absolute errors
mse = sklearn.metrics.mean_squared_error(y_test, predictions)

rmse = math.sqrt(mse)
print(rmse)
```
